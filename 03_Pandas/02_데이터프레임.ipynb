```python
import pandas as pd
```

```python
# 리스트나 행렬로 데이터프레임 생성
data1 = [
    [10, 20, 30, 40, 50],
    [100, 200, 300, 400, 500],
    [1000, 2000, 3000, 4000, 5000]
]
col1 = ['a1', 'a2', 'a3', 'a4', 'a5']
index1 = ['v1', 'v2', 'v3']
df1 = pd.DataFrame(data1, columns=col1, index=index1)
df1
```

```python
# 딕셔너리로 데이터 프레임을 생성
data2 = {
    'a1': [10, 100, 1000],
    'a2': [20, 200, 2000],
    'a3': [30, 300, 3000],
    'a4': [40, 400, 4000],
    'a5': [50, 500, 5000]
}
index2 = ['v1', 'v2', 'v3']
df2 = pd.DataFrame(data2, index=index2)
df2
```

```python
# 파일에서 데이터를 읽어온다.
df3 = pd.read_csv('iris.csv')
df3
```
